### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# configure gmaps
gmaps.configure(api_key=g_key)

# store data for heatmap
locations = city_df[["Lat", "Lng"]].astype(float)
humidity = city_df["Humidity"].astype(float)
```

```python
# create map and heatmap layer
fig = gmaps.figure(center=(28.00, 0.00), zoom_level=2)

heat_layer = gmaps.heatmap_layer(locations, weights=humidity, max_intensity = 100, point_radius = 8)

fig.add_layer(heat_layer)

fig
```
